```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = 2 * np.random.rand(100,1)
Y = 4 +3 * X+np.random.randn(100,1)
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
def calcCost(theta, X, Y):
    cost = 0
    for i in range(len(X)):
        cost = cost + ((np.sum(theta * X[i]) - Y[i]) ** 2)
    cost = cost/(2 * len(X))
    return cost
```

```python
a = np.array([[1, 5], [1, 3], [1, 3]])
b = np.array([4, 8, 12])
theta = np.array([0, 0])
calcCost(theta, a, b)
```

```python
def calcDerivative(theta, X, Y, j):
    derivative = 0
    for i in range(len(X)):
        derivative = derivative + ((np.sum(theta * X[i]) - Y[i]) * X[i][j])
    derivative = derivative/len(X)
    return derivative
```

```python
def gradientDescent(theta, X, Y, learning_rate):
    old_cost = 9999999
    new_cost = -9999999
    cost_history = np.zeros(1000)
    for i in range(1000):
        old_cost = new_cost
        temp = np.zeros(theta.shape)
        for j in range(len(theta)):
            temp[j] = theta[j] - (learning_rate * calcDerivative(theta, X, Y, j))
        theta = temp
        #print("Theta: \n", theta)
        new_cost = calcCost(theta, X, Y)
        cost_history[i] = new_cost
    print(cost_history)
    print(new_cost)
    return theta
```

```python
theta = np.array([0, 0])
learning_rate = 0.01
theta = gradientDescent(theta, X, Y, learning_rate)
```
